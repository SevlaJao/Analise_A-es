# 4. Análise univariada

```python
plt.figure(figsize = (8,6))
plt.hist(df['Preco_Abertura'], bins = 30)
plt.title('Histograma do Preço de Abertura')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.hist(df['Preco_Fechamento'], bins = 30)
plt.title('Histograma do Preço de Fechamento')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.hist(df['Menor_Preco'], bins = 30)
plt.title('Histograma do Menor preço')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.hist(df['Maior_Preco'], bins= 30)
plt.title('Histograma do Maior Preço')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.hist(df['Preco_Fechamento_Ajustado'], bins = 30)
plt.title('Histograma do Preço de Fechamento Ajustado')
plt.show()
```

### Os gráficos acima estão todos com uma assimetria para esquerda o que significa que grande parte das ações estão entre 0 - 20

# 5. Análise Bivariada

```python
Data_Fechamento = ['Data_Negociacao', 'Preco_Fechamento']
```

```python
Data_Fechamento = df[Data_Fechamento]
```

```python
Data_Fechamento = Data_Fechamento.sample(n = 4000)
```

```python
Data_Fechamento = Data_Fechamento.sort_values(by = 'Data_Negociacao')
```

```python
plt.figure(figsize = (8, 6))
plt.plot(Data_Fechamento['Data_Negociacao'], Data_Fechamento['Preco_Fechamento'])
plt.title('Linha temporal do Preço de Fechamento')
plt.xticks(rotation = 45)
plt.show()
```

```python
df['Data_Negociacao'] = df['Data_Negociacao'].astype(str)
```

```python
Data_Volume = df.groupby(df['Data_Negociacao'].apply(lambda x: x[:4]))['Volume_Negociacoes'].sum()
```

```python
plt.figure(figsize = (8, 6))
plt.barh(Data_Volume.index, Data_Volume)
plt.title('Quantidade de Negociaçoes por Ano')
plt.xlabel('Escala em bilhão')
plt.show()
```

```python
Abertura_Fechamento = ['Preco_Abertura', 'Preco_Fechamento']
```

```python
Abertura_Fechamento = df[Abertura_Fechamento]
```

```python
Abertura_Fechamento = Abertura_Fechamento.sample(n = 1000)
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(Abertura_Fechamento['Preco_Abertura'], Abertura_Fechamento['Preco_Fechamento'])
plt.title('Dispersão entre Preço de Abertura e Preço de Fechamento')
plt.xlabel('Preço de Abertura')
plt.ylabel('Preço de Fechamento')
plt.show()
```

### Aqui podemos ver que o preço das ações está aumento desde 1992 porém teve uma queda de 2023 para 2024, o ano com mais negociações da ação foi 2008 e desde esse ano até 2024 o volume de negociações está diminuindo também podemos ver que quanto maior é o preço de abertura da ação maior é o seu preço de fechamento
